```python
src = cv2.imread('images/flower.jpeg', cv2.IMREAD_GRAYSCALE)
src2 = cv2.imread('images/sky.jpeg', cv2.IMREAD_GRAYSCALE)

height, width = src.shape
src2 = cv2.resize(src2,(width,height),interpolation = cv2.INTER_AREA)

hist1 = cv2.calcHist([src], [0], None, [256], [0, 256])
hist2 = cv2.calcHist([src2], [0], None, [256], [0, 256])

sum = 0
dsum = 0
sum_of_hist = [0]*256
sum_of_dhist = [0] * 256
for i in range(256):
    sum += hist1[i,0]
    sum_of_hist[i] = sum
    dsum += hist2[i,0]
    sum_of_dhist[i] = dsum

org_temp = cv2.equalizeHist(src)

dmin = sum_of_dhist[0]
dmax = sum_of_dhist[255]
sum_of_schist = [0] * 256
for i in range(256):
    sum_of_schist[i] = int(sum_of_dhist[i] * 255 / dmax)

top = 255
bottom = top-1
table = [0]*256
while(True):
    i = sum_of_schist[bottom]
    while(i<=sum_of_schist[top]):
        table[i] = top
        i += 1
    top = bottom
    bottom -= 1
    
    if(bottom < -1):
        break

dst = np.ones_like(src)
for i in range(height):
    for j in range(width):
        dadd = int(org_temp[i,j])
        dst[i,j] = table[dadd]

hist3 = cv2.calcHist([dst], [0], None, [256], [0, 256])

dst2 = np.concatenate((src, src2, dst), axis = 1)
cv2.imshow('dst', dst2)
cv2.waitKey(0)
cv2.destroyAllWindows()
"""
fig = plt.figure(figsize = (12,5))
sub1 = fig.add_subplot(1,3,1)
sub1.plot(hist1)
sub1.set_title('src1')
sub2 = fig.add_subplot(1,3,2)
sub2.plot(hist2)
sub2.set_title('src2')
sub3 = fig.add_subplot(1,3,3)
sub3.plot(hist3)
sub3.set_title('dst')
plt.show()
"""
```

```python
threshold_img = cv2.imread("mask.jpg", cv2.IMREAD_GRAYSCALE)

element = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3), (-1,-1))
skel = np.zeros((threshold_img.shape[0],threshold_img.shape[1]), dtype=np.uint8)

while((cv2.countNonZero(threshold_img)!=0)):
    eroded = cv2.erode(threshold_img, element)
    temp = cv2.dilate(eroded, element)
    temp = cv2.subtract(threshold_img, temp)
    skel = cv2.bitwise_or(skel, temp)
    threshold_img = eroded.copy()
    
cv2.imshow("result", skel)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
net = cv2.dnn.readNet("yolov3.weights", "yolov3.cfg")
classes = []
with open("coco.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]
layer_names = net.getLayerNames()
output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers()]


def predict_yolo(img_path, bool = True):
    # 이미지 가져오기
    if bool:
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    else:
        vid = cv2.VideoCapture(0)
        bool, frame = vid.read()
        img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        vid.release()

    height, width, channels = img.shape

    blob = cv2.dnn.blobFromImage(img, 1/256, (416, 416), (0, 0, 0), swapRB=True, crop=False)
    net.setInput(blob) 
    outs = net.forward(output_layers)

    class_ids = []
    confidences = []
    boxes = []
    for out in outs:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5:
                # Object detected
                center_x = int(detection[0] * width)
                center_y = int(detection[1] * height)
                w = int(detection[2] * width)
                h = int(detection[3] * height)
                # 좌표
                x = int(center_x - w / 2)
                y = int(center_y - h / 2)
                boxes.append([x, y, w, h])
                confidences.append(float(confidence))
                class_ids.append(class_id)

    indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)

    font = cv2.FONT_HERSHEY_PLAIN
    colors = np.random.uniform(0, 255, size=(len(boxes), 3))
    if len(indexes) > 0:
        for i in indexes.flatten():
            x, y, w, h = boxes[i]
            print(x, y, w, h)
            label = str(classes[class_ids[i]])
            confidence = str(round(confidences[i], 2))
            color = colors[i]
            cv2.rectangle(img, (x, y), ((x+w), (y+h)), color, 2)
            cv2.putText(img, label + " " + confidence, (x+ 80, y-30), font, 2, (0, 255, 0), 2)
    
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        cv2.imshow("obj detec",img)
        cv2.waitKey()
        cv2.destroyAllWindows()
  
    else:
        print('탐지된 물체가 없습니다.')
    
    
predict_yolo("vid_4_1000.jpg", False)
```
